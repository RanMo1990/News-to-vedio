```python
import re

def remove_section_headers_emojis_and_quotes(script):
    # Split script into lines
    lines = script.split("\n")

    # Remove lines that contain section headers
    filtered_lines = [line for line in lines if not line.strip().startswith("üé¨ **Hook (") 
                      and not line.strip().startswith("üé¨ **Body (") 
                      and not line.strip().startswith("üé¨ **Closing (")]

    # Join the filtered lines back into a single text
    cleaned_script = "\n".join(filtered_lines)

    # Remove all emojis using regex (matches any Unicode emoji character)
    cleaned_script = re.sub(r"[^\w\s.,!?‚Äò‚Äô]", "", cleaned_script)  # Keeps only words, spaces, punctuation

    # Remove all types of quotation marks
    cleaned_script = re.sub(r"[‚Äú‚Äù\"']", "", cleaned_script)

    return cleaned_script


cleaned_script = remove_section_headers_emojis_and_quotes(video_script)
print(cleaned_script)
```

```python
import os
import subprocess
import re
from gtts import gTTS
from pydub import AudioSegment
from PIL import Image, ImageDraw, ImageFont



# Manually specify FFmpeg path if not in system PATH
AudioSegment.converter = r"C:\ProgramData\chocolatey\bin\ffmpeg.exe"
AudioSegment.ffprobe = r"C:\ProgramData\chocolatey\bin\ffprobe.exe"

# ‚úÖ Set FFmpeg Path (Change if needed)
FFMPEG_PATH = r"C:\ProgramData\chocolatey\bin\ffmpeg.exe"  # Use full path

# ‚úÖ Function to split script into sentences
def split_script(script):
    # List of common abbreviations that should not cause a split
    abbreviations = {"U.S.", "Dr.", "Mr.", "Mrs.", "Ms.", "Inc.", "Ltd.", "Jr.", "Sr.", "e.g.", "i.e."}

    # Temporary replacements for abbreviations to prevent incorrect splits
    temp_abbr = {abbr: abbr.replace(".", "<DOT>") for abbr in abbreviations}
    for abbr, replacement in temp_abbr.items():
        script = script.replace(abbr, replacement)

    # Split sentences correctly (but not inside protected abbreviations)
    sentences = re.split(r'(?<=[.!?])\s+', script)

    # Restore abbreviations
    sentences = [s.replace("<DOT>", ".").strip() for s in sentences if s.strip()]

    return sentences

# ‚úÖ Function to generate speech and measure duration
def text_to_speech(sentences, output_audio="voice.mp3"):
    total_audio = AudioSegment.silent(duration=0)
    sentence_durations = []

    for i, sentence in enumerate(sentences):
        tts = gTTS(sentence, lang="en")
        temp_audio_file = f"temp_{i}.mp3"
        tts.save(temp_audio_file)

        # Load the audio to measure duration
        audio = AudioSegment.from_file(temp_audio_file)
        duration = len(audio) / 1000  # Convert ms to seconds
        sentence_durations.append(duration)

        total_audio += audio + AudioSegment.silent(duration=500)  # Add a small pause

        os.remove(temp_audio_file)  # Cleanup

    total_audio.export(output_audio, format="mp3")
    print(f"üé§ Audio saved as: {output_audio}")
    
    return output_audio, sentence_durations

# ‚úÖ Function to create FFmpeg file list for correct timing
def create_file_list(image_folder, durations, file_list="file_list.txt"):
    images = sorted(os.listdir(image_folder))  # Sort to match order
    images = [img for img in images if img.endswith(('.png', '.jpg', '.jpeg'))]

    if len(images) != len(durations):
        print(images)
        print(durations)
        print("‚ö†Ô∏è Warning: Number of images does not match sentences! Adjust manually.")
    
    with open(file_list, "w") as f:
        for img, duration in zip(images, durations):
            f.write(f"file '{os.path.join(image_folder, img)}'\n")
            f.write(f"duration {duration}\n")
    
    return file_list

# ‚úÖ Function to create video with correct durations
def create_video(image_folder, audio_file, durations, output_video="output.mp4"):
    file_list = create_file_list(image_folder, durations)

    # Step 1: Create Video from Images
    temp_video = "temp_video.mp4"
    subprocess.run([
        FFMPEG_PATH, "-y", "-f", "concat", "-safe", "0",
        "-i", file_list,
        "-vsync", "vfr", "-pix_fmt", "yuv420p",
        temp_video
    ], check=True)

    print("üé¨ Temporary video from images created.")

    # Step 2: Merge Video with Audio
    subprocess.run([
        FFMPEG_PATH, "-y",
        "-i", temp_video,
        "-i", audio_file,
        "-c:v", "copy",
        "-c:a", "aac",
        "-shortest", output_video
    ], check=True)

    print(f"‚úÖ Final video created: {output_video}")

# ‚úÖ Run the Full Process
if __name__ == "__main__":
    # üìù Define the text script
    script_text = cleaned_script

    # üìú Split script into sentences
    sentences = split_script(script_text)
    print(sentences)
    # üé§ Convert script to speech & get durations
    audio_path, durations = text_to_speech(sentences, "voice.mp3")

    # üé¨ Create a video using images + audio
    create_video("google_images", audio_path, durations, "final_video.mp4")
```
